```python
# Trend Over Time
```

```python
# Create sample data

# 'days' represents time (x-axis)
days = [1, 2, 3, 4, 5]

# 'temperature' represents values (y-axis)
temperature = [30, 32, 35, 33, 31]
```

```python
# Create a new figure with a specific size (width=7 inches, height=4 inches)
plt.figure(figsize=(7, 4))

# Plot a line graph
# marker='o' adds circular points on each data value
plt.plot(days, temperature)

# Add a title to the graph
plt.title("Temperature Trend")

# Label the x-axis
plt.xlabel("Days")

# Label the y-axis
plt.ylabel("Temperature (°C)")

# Show grid lines for better readability
plt.grid(True)

# Display the graph
plt.show()
```

```python
# 8. Pie Chart
# Used to show proportions.
```

```python
# Data
labels = ["A", "B", "C"]
sizes = [30, 40, 30]

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Add title (optional)
plt.title("Simple Pie Chart")

# Display plot
plt.show()

# autopct (short for "automatic percentage") is a parameter used in the plt.pie() function 
# to display numerical values—typically percentages—directly on the slices of a pie chart.

# autopct='%1.1f%%': Displays the percentage with one decimal place (e.g., 15.5%).
# autopct='%1.0f%%': Displays the percentage as a whole number (e.g., 16%).
# autopct='%.2f': Displays the percentage with two decimal places but no "%" sign.
```

```python
# Histogram (Frequency Distribution)
```

```python
# A histogram divides data into intervals (bins) and counts how many values fall into each bin.

# It helps you understand:
# Shape (normal, skewed, uniform)
# Spread (wide vs narrow)
# Concentration of values
```

```python
import matplotlib.pyplot as plt

scores = [42, 45, 48, 52, 55, 58, 62, 65, 70, 75]

# 1. Capture 'bins' (the edges) and 'patches' (the bars)
counts, bins, patches = plt.hist(scores, bins=5, edgecolor='black', color='skyblue')

# 2. Force the x-axis to show the exact interval numbers
plt.xticks(bins)

# 3. Add the count labels on top of each bar
plt.bar_label(patches)

plt.title("Histogram of Student Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")

plt.show()
```

```python
# Box Plot (Statistical Summary + Outliers)
```

```python
# A Box Plot is a visual summary of data that shows how your numbers are spread 
# out by dividing them into four equal parts.

# Minimum: The lowest value in the data.
# Q1 (First Quartile): The point where the bottom 25% of the data ends.
# Median: The exact middle value of the data.
# Q3 (Third Quartile): The point where the bottom 75% of the data ends.
# Maximum: The highest value in the data.

# # Also highlights outliers
```

```python
scores = [45, 45, 48, 52, 55, 58, 62, 65, 70, 75]

# Use tick_labels for the x-axis label
plt.boxplot(scores, tick_labels=['Student Scores'])

# Set the y-axis limits (bottom, top)
plt.ylim(40, 80) 

plt.title("Box Plot of Student Scores")
plt.ylabel("Scores")
plt.show()
```

```python
# Imagine these 10 students are lined up from lowest to highest score:

# 1. Minimum: 45
# The very first score in the line. It's the lowest grade anyone got.

# 2. Q1 (25% mark): 48
# If you look at the bottom half of the class, 48 is the middle of that group. 
# It means 25% of students scored 48 or lower.

# 3. Median (50% mark): 56.5
# This is the middle of the whole line. Since you have 10 students, we take the average of the two middle scores (55 and 58).
# Meaning: Half the class scored above 56.5, and half scored below.

# 4. Q3 (75% mark): 65
# If you look at the top half of the class, 65 is the middle of that group. 
# It means 75% of the class scored 65 or lower.

# 5. Maximum: 75
# The highest score in the class. The "winner" of the test.
```

```python
# Scatter Plot
```

```python
# A scatter plot places one variable on the x-axis and another on the y-axis. 

# Each point represents a single observation. By looking at the pattern of points, you can quickly assess:
# Positive correlation → points trend upward (as x increases, y increases)
# Negative correlation → points trend downward (as x increases, y decreases)
# No correlation → points are randomly scattered
# Clusters or groups → potential subcategories in data
# Outliers → unusual data points that don’t follow the pattern
```

```python
# Sample data
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 4, 5, 4, 5, 7, 8]

# Create scatter plot
plt.scatter(x, y)

# Labels and title
plt.xlabel("X Variable")
plt.ylabel("Y Variable")
plt.title("Scatter Plot Example")

plt.show()
```
